```python
# Add the parent directory of the current working directory to the Python path at runtime. 
# In order to import modules from the src directory.
import os
import sys 
from dotenv import load_dotenv, find_dotenv

current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
sys.path.insert(0, parent_dir)

load_dotenv(find_dotenv())

BASEPATH = os.environ["BASEPATH"]
```

```python
import pandas as pd
import numpy as np
```

```python
CSV_PATH = os.path.join(BASEPATH, "Internal/Intermediate/CSV")
articles_file = os.path.join(BASEPATH, "Public/Data/CSV/neuroscience_articles_1999-2023.csv")

neuroscape_articles = pd.read_csv(articles_file)
```

```python
multidisciplinary_new_file = os.path.join(CSV_PATH, "Multidisciplinary/merged_new.csv")
multidisciplnay_old_file = os.path.join(CSV_PATH, "Multidisciplinary/merged_old.csv")
combined_file = os.path.join(CSV_PATH, "Multidisciplinary/articles_merged_cleaned.csv")

md_new = pd.read_csv(multidisciplinary_new_file)
md_old = pd.read_csv(multidisciplnay_old_file)

md_combined = pd.concat([md_old, md_new]).drop_duplicates(subset=["Pmid"]).reset_index(drop=True)

# drop rows where Pmid is already in neuroscape_articles
md_combined = md_combined[~md_combined["Pmid"].isin(neuroscape_articles["Pmid"])].reset_index(drop=True)

md_combined.to_csv(combined_file, index=False)
```

```python
neuroscience_new_file = os.path.join(CSV_PATH, "Neuroscience/merged_new.csv")
neuroscience_old_file = os.path.join(CSV_PATH, "Neuroscience/merged_old.csv")
combined_file = os.path.join(CSV_PATH, "Neuroscience/articles_merged_cleaned.csv")

neuro_new = pd.read_csv(neuroscience_new_file)
neuro_old = pd.read_csv(neuroscience_old_file)

neuro_combined = pd.concat([neuro_old, neuro_new]).drop_duplicates(subset=["Pmid"]).reset_index(drop=True)

# drop rows where Pmid is already in neuroscape_articles
neuro_combined = neuro_combined[~neuro_combined["Pmid"].isin(neuroscape_articles["Pmid"])].reset_index(drop=True)

neuro_combined.to_csv(combined_file, index=False)
```

```python
neuroscience_old_file = os.path.join(CSV_PATH, "Neuroscience/articles_merged_cleaned_filtered_old.csv")
combined_file = os.path.join(CSV_PATH, "Neuroscience/articles_merged_cleaned_filtered.csv")

neuro_old = pd.read_csv(neuroscience_old_file)

neuro_combined = pd.concat([neuro_old, neuroscape_articles]).drop_duplicates(subset=["Pmid"]).reset_index(drop=True)

neuro_combined.to_csv(combined_file, index=False)
```
